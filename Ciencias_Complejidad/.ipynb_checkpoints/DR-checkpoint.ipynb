```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from matplotlib import animation
plt.rcParams['animation.embed_limit'] = 2**128
from IPython import display
```

```python
A = np.zeros((20, 20))
```

```python
A[10, 10] = 1
plt.imshow(A)
```

```python
w = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
    ])
```

```python
r = 0.1
B = A + r*(signal.correlate2d(A, w, mode='same'))
#print(B)
plt.imshow(B)
```
